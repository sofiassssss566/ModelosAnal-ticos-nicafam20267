**2. Escribe código**

Dado peso = 70 (kg) y altura = 1.75 (m), calcula el IMC (peso / altura²) e imprime la categoría: Bajo peso (<18.5), Normal (18.5–24.9), Sobrepeso (25–29.9), Obesidad (≥30).

```python
peso = 70
altura = 1.75

imc = peso / (altura ** 2)

print("IMC:", round(imc, 2))

if imc < 18.5:
    print("Categoría: Bajo peso")

elif imc < 25:
    print("Categoría: Normal")

elif imc < 30:
    print("Categoría: Sobrepeso")

else:
    print("Categoría: Obesidad")
```

```python
monto = 120000

if monto < 50000:
    descuento = 0

elif monto < 100000:
    descuento = monto * 0.10

elif monto < 200000:
    descuento = monto * 0.20

else:
    descuento = monto * 0.30

total = monto - descuento

print("Monto de la compra:", monto)
print("Descuento:", descuento)
print("Total a pagar:", total)
```

**6. Escribe código**

Para un número n, imprime: "FizzBuzz" si es múltiplo de 3 y de 5; "Fizz" si es múltiplo solo de 3; "Buzz" si es múltiplo solo de 5; el número si no cumple ninguna. Prueba con n = 15, n = 9, n = 10 y n = 7.

```python
numeros = [15, 9, 10, 7]

for n in numeros:

    if n % 3 == 0 and n % 5 == 0:
        print(n, "-> FizzBuzz")

    elif n % 3 == 0:
        print(n, "-> Fizz")

    elif n % 5 == 0:
        print(n, "-> Buzz")

    else:
        print(n)
```

**2. Escribe código**

Usando un ciclo for, suma todos los números del 1 al 100 e imprime el resultado. Luego suma solo los números pares del mismo rango.

```python
# Suma de 1 a 100

suma = 0

for i in range(1, 101):
    suma += i

print("La suma de 1 a 100 es:", suma)

# Suma de los números pares

suma_pares = 0

for i in range(2, 101, 2):
    suma_pares += i

print("La suma de los números pares es:", suma_pares)
```

**3. Escribe código**

Dada precios = [12000, 45000, 8500, 67000, 3200], usa un ciclo para: (1) imprimir cada precio con formato Item 1: $12,000, (2) imprimir al final cuántos artículos cuestan más de $10,000.

```python
precios = [12000, 45000, 8500, 67000, 3200]

contador = 0

for i in range(len(precios)):
    print(f"Item {i+1}: ${precios[i]:,}")

    if precios[i] > 10000:
        contador += 1

print("Artículos que cuestan más de $10.000:", contador)
```

**4. Escribe código**

Un banco tiene saldo = 1000. Cada mes retiras 150 y se cobran 20 de comisión. Usa un ciclo while para simular cuántos meses pasan antes de que el saldo llegue a cero o menos. Imprime el saldo al final de cada mes

```python
saldo = 1000
mes = 0

while saldo > 0:
    saldo = saldo - 150 - 20
    mes += 1

    print("Mes", mes, "- Saldo:", saldo)

print("El saldo llegó a cero o menos después de", mes, "meses.")
```

**5. Escribe código**

Dada la lista numeros = [3, 7, -2, 15, 4, -8, 9, 1], escribe un ciclo que: (1) saltee los números negativos con continue, (2) se detenga con break al encontrar un número mayor a 10, (3) imprima los números que procesa.

```python
numeros = [3, 7, -2, 15, 4, -8, 9, 1]

for numero in numeros:

    if numero < 0:
        continue

    if numero > 10:
        print("Se encontró un número mayor a 10:", numero)
        break

    print(numero)
```
