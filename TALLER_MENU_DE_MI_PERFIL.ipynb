```python
# Taller: Menú de mi Perfil con Diccionarios en Python
```

```python
perfil = {
    "nombre": "Yinet Sofía Sánchez",
    "edad": 19,
    "ciudad": "Bogotá",
    "institucion": "CAFAM",
    "estado_civil": "Soltera",
    "hobbies": [
        "Escuchar música",
        "Ver películas",
        "Compartir con amigos"
    ]
}
```

```python
print("\n========== MENÚ ==========")
print("1. Agregar un dato nuevo")
print("2. Actualizar un dato existente")
print("3. Eliminar un dato con del")
print("4. Eliminar un dato con pop()")
print("5. Consultar un dato con get()")
print("6. Ver el perfil completo")
```

```python
opcion = input("\nSeleccione una opción (1-6): ")
```

```python
if opcion == "1":
    clave = input("Ingrese la nueva clave: ")
    valor = input("Ingrese el valor correspondiente: ")

    perfil[clave] = valor

    print("\nDato agregado correctamente.")
    print(perfil)
```

```python
opcion = input("\nSeleccione una opción (1-6): ")
```

```python
if opcion == "2":
    clave = input("Ingrese la clave que desea actualizar: ")

    if clave in perfil:
        nuevo_valor = input("Ingrese el nuevo valor: ")
        perfil[clave] = nuevo_valor

        print("\nDato actualizado correctamente.")
        print(perfil)
    else:
        print("\nLa clave ingresada no existe.")
```

```python
opcion = input("\nSeleccione una opción (1-6): ")
```

```python
if opcion == "3":
    clave = input("Ingrese la clave que desea eliminar: ")

    if clave in perfil:
        del perfil[clave]

        print("\nDato eliminado correctamente.")
        print(perfil)
    else:
        print("\nLa clave ingresada no existe.")
```

```python
opcion = input("\nSeleccione una opción (1-6): ")
```

```python
if opcion == "4":
    clave = input("Ingrese la clave que desea eliminar: ")

    if clave in perfil:
        dato_eliminado = perfil.pop(clave)

        print("\nEl valor eliminado fue:", dato_eliminado)
        print(perfil)
    else:
        print("\nLa clave ingresada no existe.")
```

```python
opcion = input("\nSeleccione una opción (1-6): ")
```

```python
if opcion == "5":
    clave = input("Ingrese la clave que desea consultar: ")

    resultado = perfil.get(clave, "No disponible")

    print("\nResultado de la consulta:", resultado)
```

```python
opcion = input("\nSeleccione una opción (1-6): ")
```

```python
if opcion == "6":
    print("\n===== PERFIL COMPLETO =====")
    print(perfil)
```
